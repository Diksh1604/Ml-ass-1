# **Ans 15:-**

Causation refers to relationship between phenomena where one event directly influences the occurrence of the other effect. This relationship is often tested through research that establish a direct link between cause and effect.

### **Difference Between Correlation and Causation**

**Correlation:**

Correlation is a statistical relationship between variables where tend vary together.

**Example:**

 Ice cream sales and the number of people who go swimming. Both tend to increase during the summer months. However, this does not mean that eating ice cream causes people to go swimming or vice versa. The underlying factor is likely the warm weather, which drives both behaviors.

**Caustion:**

It says that one event or variable directly causes another to happen.

**Example:-**

Drinking too much alcohol can cause serious health issues.

Optimizer is an algorithm which is used to minimize or maximize a function with adjusting the values of there parameters.

### **Types of Optimizer:**

**1)Adam:** Combines momentum and adaptive learning rates.

**Example:-** Go-to optimizer for modern architectures.



**2)Mini-batch SGD:** Compromise used in practice.

**Example:-** If you have a dataset with 1,000 samples, a mini-batch size of 100 would result in 10 updates per iterate commonly used in deep learning.

**3)Nesterov:** Improved version of momentum.

**Example:-** Advanced optimization tasks where quick convergence is needed, like training recurrent neural networks.

**4)Stochastic Gradient Descent:**  Updates the parameters using a particular example.

**Example:-** Computing the gradient for all training samples, computes the gradient for randomly chosen training sample and updates the weights accordingly.

**5)Gradient Descent:** Basic and works well with few datasets.

**Example:-** Gradient descent will update the weights and biases by parameters.

**6)Momentum:** Helps in accelerate convergence.

**Examples:-** Optimization path is oscillating or stuck in a shallow region, the momentum term helps to push the updates in a consistent direction.

**7)RMSprop:** Stabilizes gradient descent.

**8)Adagrad:** Learning rate of each parameter.

# **Ans 21:-**

Feature scaling is the process of transforming the features of dataset into a similar scale or range, it ensures that no single feature dominates to learning process, otherwise negative affect performance of machine learning algorithms.

 **Helps to feature scaling in Machine Learning:**

 1) Avoids Dominance in Distance-Based Models
 2) Improves Convergence in Gradient-Based Algorithms
 3) Improves Performance in Regularization
 4) Ensures Equal Contribution of Features
